```python
import pandas as pd
import numpy as np
import math
```

```python
def nd(x, mean,var):
    exponent = math.exp(-(math.pow(x - mean, 2) / (2 * math.pow(var, 2))))
    return (1 / (math.sqrt(2 * math.pi) * var)) * exponent



def predict(test, data_means, data_var):
    pred0 = list(range(0,len(test)))
    pred1 = list(range(0,len(test)))
    pred = list(range(0,len(test)))
    
    for i in range(0,len(test)):
        for j in range(0,8):
            pred0[i] *= nd(test[j][i], data_means[j][0], data_var[j][0])
            pred1[i] *= nd(test[j][i], data_means[j][1], data_var[j][1])
       
    for i in range(len(test)):
        if(pred0[i] >= pred1[i]):
            pred[i] = 0
        else:
            pred[i] = 1
    
    return pred

def accuracy(prediction):
    count = 0
    for i in range(0,len(test)):
        if(test[8][i] == prediction[i]):
            count+=1
    accuracy = (count / len(test)) * 100
    return accuracy



def confusionMat(prediction):
    TN = 0
    TP = 0
    FN = 0
    FP = 0
    
    for i in range(0,len(test)):
        if((test[8][i] == 0) and (prediction[i] == 0)):
            TN +=1
        elif((test[8][i] == 0) and (prediction[i] == 1)):
            FP +=1
        elif((test[8][i] == 1) and (prediction[i] == 0)):
            FN +=1
        elif((test[8][i] == 1) and (prediction[i] == 1)):
            TP +=1
    
    acc =  (TP + TN) / (TP + FP + TN + FN) * 100
    err = (FP + FN) / (TP + FP + TN + FN) * 100
    Sensitivity = TP / (FN + TP)
    Specificity = TN / (TN/FP)
    print(f"Accuracy using confusion matrix = {acc}")
    print(f"Error using confusion matrix = {err}")
    #sensitivity 
    print(f"Sensitivity using confusion matrix = {Sensitivity}")
    print(f"Specificity using confusion matrix = {Specificity}")

def main():
    train = pd.read_csv("train.csv", header = None)
    test = pd.read_csv("test.csv", header = None)
    
    n_0 = train.iloc[:,8][train.iloc[:,8] == 0].count()
    n_1 = train.iloc[:,8][train.iloc[:,8] == 1].count()
    #print(f"Number of 0 = {n_0}      Number of 1 = {n_1}")
    total = train.iloc[:,8].count()
    #print(f"total = {total}")
    P_0 = n_0/total
    P_1 = n_1/total
    #print(f"P0 = {P_0}     P1 = {P_1}")
    
    data_means = train.iloc[:,0:8].groupby(train.iloc[:,8]).mean()
    data_var = train.iloc[:,0:8].groupby(train.iloc[:,8]).std()
    
    prediction = predict(test,data_means, data_var)
    
    acc = accuracy(prediction)
    print(f"Accuracy = {acc}")
    #print(prediction)
    confusionMat(prediction)

if __name__ == "__main__" : main()
```
